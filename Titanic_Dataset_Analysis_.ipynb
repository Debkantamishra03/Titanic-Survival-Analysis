```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python


# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")
gender_submission = pd.read_csv("/content/gender_submission.csv")
```

```python
# --- Data Cleaning ---
train_cleaned = train_df.drop(columns=["Name", "SibSp", "Parch", "Embarked", "Cabin"])
train_cleaned = train_cleaned.dropna()

# --- Analysis ---
# Survival counts
print("Total survived:", train_cleaned['Survived'].sum())
print("Male survived:", train_cleaned['Survived'][train_cleaned['Sex'] == "male"].sum())
print("Female survived:", train_cleaned['Survived'][train_cleaned['Sex'] == "female"].sum())

# Survival by age
avg_age = train_cleaned['Age'].mean()
print("Older survived:", train_cleaned['Survived'][train_cleaned['Age'] >= avg_age].sum())
print("Younger survived:", train_cleaned['Survived'][train_cleaned['Age'] < avg_age].sum())

# Fare stats
print("Average fare:", train_cleaned['Fare'].mean())
print("Max fare:", train_cleaned['Fare'].max())
print("Min fare:", train_cleaned['Fare'].min())
```
